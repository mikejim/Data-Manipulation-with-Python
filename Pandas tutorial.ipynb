**Pandas Series**
1D arrays with axis labels

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html

```python
import pandas as pd
print(pd.__version__)
```

```python
A = pd.Series([2,3,4,5], index=['a', 'b', 'c', 'd'])
print(A.values)
print (type(A))
print (type(A.values))
print(A['a':'c'])
```

```python
dict = {'A':4, 'A-': 3.6, 'B': 3.3, 'B-': 3.0, 'C': 2.5, 'D': 2.0}
grades = pd.Series(dict)
print(grades)

dict2 = {'A':84, 'A-':80, 'B': 75, 'C': 65, 'D-': 55}
marks = pd.Series(dict2)
print(marks)
```

```python
#Indices

A = pd.Series(['a', 'b', 'c'], index=[1,3,5])
A[1]

print(A[1:3])

#Explicit indices
print(A.loc[1:3])

#Implicit indices
print(A.iloc[1:3])
```

**DataFrames**

```python
D = pd.DataFrame({'Marks': marks, 'Grades:': grades})
print(D)
print(D.T)

print(D.values[2,0])

print(D.columns)

D['ScaledMarks'] = (D['Marks']/90)*100
print(D)

del D['ScaledMarks']
print(D)

G = D[D['Marks'] > 70]
print(G)
```

```python
A = pd.DataFrame([{'a': 1, 'b': 4}, {'b': 3, 'c': 5}])
print(A)

#Filla NaN with zeros
B = A.fillna(0)
print(B)

#Will drop NaN values, thus deleting all the rows
#A.dropna()
```
