**Manipulating Excel Files**

```python
def convert_people_cell(cell):
  if cell == 'not available':
    return 'Diaz'
  return cell

def convert_name_cell(cell):
  if cell == 'not available':
    return 'Magdalena'
  return cell

df = pd.read_excel("/content/drive/My Drive/Colab Notebooks/Python Tutorials/Names.xlsx", 'Hoja1', converters={
    'LastName': convert_people_cell,
    'Name': convert_name_cell
})
df

#This creates an offset at which the columns and rows will be added
df.to_excel('/content/drive/My Drive/Colab Notebooks/Python Tutorials/My_File.xlsx', sheet_name='laPeople', startrow=1, startcol=2, index=False)
```

How to write an Excel with two sheets

```python
namesAges = pd.DataFrame({'Name':['Tom', 'Jack', 'Steve', 'Ricky'],
             'Age':[28,34,29,42]})
namesDoB  = pd.DataFrame({'Name':['Miguel', 'Ana', 'Helicondia', 'Ricky'],
             'DoB':['1/1/2019','1/1/1999', '4/4/1998', '2/3/2001']})

filePath = '/content/drive/My Drive/Colab Notebooks/Python Tutorials/ExcelTwoSheets.xlsx'

with pd.ExcelWriter(filePath) as writer:
  namesAges.to_excel(writer, sheet_name='Ages')
  namesDoB.to_excel(writer, sheet_name='DoB')
```
