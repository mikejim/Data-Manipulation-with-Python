## **Dataframe Tutorial**

As followed in Youtube tutorial: https://www.youtube.com/watch?v=F6kmIpWWEdU

```python
import pandas as pd
weatherdata = {
    'day': ['1/1/2021', '1/2/2021', '1/3/2021', '1/4/2021', '1/5/2021', '1/6/2021'],
    'temperature': [33, 22, 31, 28, 29, 28],
    'windspeed': [4, 5, 6, 3, 9, 7],
    'event': ['Rain', 'Snow', 'Snow', 'Sunny', 'Sunny', 'Snow']
}
df = pd.DataFrame(weatherdata)
print(df)

print(df.day)
print(df['event'])
print(df[['event', 'day']])
```

```python
#Calculate maximum of a column

print(df['temperature'].max())
print(df['temperature'].min())
print(df['temperature'].std())
print(df.describe())
```

```python
#Conditions

print(df[df.temperature >=32])

print(df[df.temperature == df.temperature.max()])

print(df[['day', 'temperature']][df.temperature == df.temperature.max()])
```
